```python
from pyts.image import GramianAngularField
```

```python
t = np.linspace(0,4*np.pi,1000)
```

```python
x = np.sin(t)
plt.plot(x)
```

```python
methods = ['summation' , 'difference' ]
```

```python
gadf = GramianAngularField(method='summation')
Xs = gadf.fit_transform(np.array([x]))
```

```python
Xs.shape
```

```python
plt.imshow(Xs[0])
```

```python
def GAF(X):

    gasf1 = GramianAngularField(method='summation')
    Xs = gasf1.fit_transform(np.array([X]))

    gadf1 = GramianAngularField(method='difference')
    Xd = gadf1.fit_transform(np.array([X]))

    fig = plt.figure(figsize=(10,10))

    ax1= fig.add_subplot(2,1,1)
    ax1.plot(X)
    ax2= fig.add_subplot(2,2,3)
    ax2.imshow(Xs[0],cmap='jet')

    ax3= fig.add_subplot(2,2,4)
    ax3.imshow(Xd[0],cmap='jet')
```

```python
GAF(x)
```

```python
GAF(t)
```

```python
GAF(np.cos(t)**2 - np.sin(t)**3)
```

```python
GAF(np.cos(t)**5 - np.sin(t)**3 )
```

```python
GAF(np.exp(t))
```
